```python
x = ['Apples', 'Bananas', 'Cherries', 'Dates', 'Elderberries', 
     'Figs', 'Grapes', 'Honeydew', 'Kiwi', 'Lemons']
y = [48, 38, 24, 17, 30, 48, 28, 32, 20, 20]
```

```python
plt.figure(figsize = (10,6))
plt.bar(x,y , color = 'maroon' , edgecolor = 'red')
```

```python
plt.figure(figsize = (8,5))
plt.barh(x,y , color = 'maroon' , edgecolor = 'red')
```

```python
y2 = [49, 18, 15, 38, 32, 32, 16, 51, 30, 19]

lenn = np.arange(len(x))
width = 0.4

plt.figure(figsize= (8,8))
plt.bar(lenn,y , width = width)
plt.bar (lenn+width, y2 , width = width)
```
